```python
CreditRisk['person_emp_length'].fillna(CreditRisk['person_emp_length'].median(), inplace=True)
CreditRisk['loan_int_rate'].fillna(CreditRisk['loan_int_rate'].mean(), inplace=True)
```

individuals employment length is heavily influenced by outliers, so it is better to replace the missing values by the median. and since mean values are sensitive to outliers where the loan interest rate have low outliers it's best to replace the missing values by the mean
